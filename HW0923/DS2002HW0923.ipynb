1. Reading the CSV File:
- Load the provided CSV file into a pandas DataFrame.
- Display the first 5 rows of the DataFrame to understand the structure of the data.

```python
import pandas as pd

# Load CSV file
df = pd.read_csv('acc_players-2324F.csv')

# Display first 5 rows
display(df.head())
```

2. Basic Analysis:
- Calculate the total points scored by all players combined.
- Find the player who has played the most minutes (MP).
- Identify the top 5 players in terms of total rebounds (TRB).

```python
# Calculate the total points scored
totalpts = df['PTS'].sum()
print(totalpts)
```

```python
# Player who played the most minutes
most_minutes = df[['Player', 'MP']].sort_values(by = 'MP', ascending = False).head(1)
display(most_minutes)
```

```python
# Top 5 in rebounds
top5rbds = df[['Player', 'TRB']].sort_values(by = 'TRB', ascending = False).head(5)
display(top5rbds)
```

3. Player Filtering:
- Create a new DataFrame containing only players who played more than 500 minutes.
- From this filtered DataFrame, determine the player with the highest total assists (AST).
- Who are the top 3 Assist Leads in the League?
- Who are the top 3 Shot Blockers?

```python
# Filtered DF
dfnew = df.sort_values(by = 'MP', ascending = False).head(500)

# Highest total assists
highestassists = dfnew[['Player', 'AST']].sort_values(by = 'AST', ascending = False).head(1)
display(highestassists)
```

```python
# Top 3 assist leaders
top3assists = dfnew[['Player', 'AST']].sort_values(by = 'AST', ascending = False).head(3)
display(top3assists)
```

```python
# Top 3 shot blockers
top3sb = dfnew[['Player', 'BLK']].sort_values(by = 'BLK', ascending = False).head(3)
display(top3sb)
```

4. School-Based Analysis:
- Group the players by School and calculate the total points scored by each school.
- Group the players by School and calculate the total assists (AST) for each team.
- Sort the schools by total points scored and display the top 3 schools.

```python
# Total points by school
totalpts = df.groupby('School')[['PTS']].sum().sort_values(by = 'PTS', ascending = False).reset_index()
display(totalpts)
```

```python
# Total assists by school
totalast = df.groupby('School')[['AST']].sum().sort_values(by = 'AST', ascending = False).reset_index()
display(totalast)
```

```python
# Top 3 schools in pts
display(totalpts.head(3))
```
